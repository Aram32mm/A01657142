```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


#Carga los datos usando tu lector de csv o con pandas. Es recomendable hacerlo con pandas.
data = pd.read_csv("avocado.csv")
print(data.head())

#Realiza el an√°lisis de las variables usando diagramas de cajas y bigotes, histogramas y mapas de calor.
```

```python
#----------Bigotes----------

#General
boxplot = data.boxplot()
boxplot.plot()
plt.show()

#Average Price
boxplot = data.boxplot(column=['AveragePrice'])
boxplot.plot()
plt.show()

#Total Volume
boxplot = data.boxplot(column=['Total Volume'])
boxplot.plot()
plt.show()

#4046
boxplot = data.boxplot(column=['4046'])
boxplot.plot()
plt.show()

#4225
boxplot = data.boxplot(column=['4225'])
boxplot.plot()
plt.show()

#4770
boxplot = data.boxplot(column=['4770'])
boxplot.plot()
plt.show()

#Total Bags
boxplot = data.boxplot(column=['Total Bags'])
boxplot.plot()
plt.show()

#Small Bags
boxplot = data.boxplot(column=['Small Bags'])
boxplot.plot()
plt.show()

#Large Bags
boxplot = data.boxplot(column=['Large Bags'])
boxplot.plot()
plt.show()

#XLarge Bags
boxplot = data.boxplot(column=['XLarge Bags'])
boxplot.plot()
plt.show()

#year
boxplot = data.boxplot(column=['year'])
boxplot.plot()
plt.show()
```

```python
#----------Histograma----------
##Op 1:
plt.style.use('ggplot')
plt.hist(data['AveragePrice'], bins=10)
plt.show()

##Op 2:
histogram = data.plot.hist()
print(histogram)
plt.show()

tipos = data.dtypes                      
print(tipos)
```
